```python
import volcaniQ as vq
import numpy as np
import matplotlib.pyplot as plt
```

```python
dm = vq.DryModulus(mineral_modulus=36, porosity=0.2)
phi = np.arange(0,1,.01)
dry_mod = [dm(p).mavko_mukerji_dry_modulus(pore_space_modulus=10) for p in phi]
dry_mod2 = [dm(p).mavko_mukerji_dry_modulus(pore_space_modulus=5) for p in phi]
dry_mod3 = [dm(p).mavko_mukerji_dry_modulus(pore_space_modulus=1) for p in phi]
plt.plot(phi, dry_mod)
plt.plot(phi, dry_mod2)
plt.plot(phi, dry_mod3)
plt.show()
plt.close()
```
